```python
sn = []
st = []
si = []


savings = {}


driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
driver.implicitly_wait(3)
# url = 'https://obank.kbstar.com/quics?page=C103429' # 대출 페이지 
# driver.get(url)
# time.sleep(3)

# # chrome driver
# #driver = webdriver.Chrome(service= service, options=chrome_options) # <- options로 변경

# # kb 국민은행 
# url = 'https://obank.kbstar.com/quics?page=C016613&cc=b061496:b061496' # 적금 페이지 
url = 'https://obank.kbstar.com/quics?page=C016613&cc=b061496:b061496'

# https://obank.kbstar.com/quics?page=C016613&cc=b061496:b061496#none

driver.get(url)
time.sleep(3)

driver.find_element(By.XPATH, '//*[@id="categoryD00027"]').click() # 적금 페이지
time.sleep(1)
savings_nums = driver.find_elements(By.CLASS_NAME, 'total_num')
nums = int(re.sub(r'[^0-9]', '', savings_nums[0].text))  # 상품 수
if nums % 10 ==0 : # 페이지 수 
    nums = nums // 10
else: 
    nums = (nums // 10) + 1 



print(nums)
for page in range(nums):
    page+=1

    driver.find_element(By.CSS_SELECTOR, f"[id='pageinput{page}']").click()
    time.sleep(3)
    savings_name = driver.find_elements(By.CLASS_NAME, 'title') # 예금 상품 제목 가져오기 
    for i in range(len(savings_name)): 
        
        
        i += 1
        # print(f'상품번호{i}')
        # print(f'//*[@id="b061496"]/ul[2]/li[{i}]/div[1]/a')
        driver.find_element(By.XPATH, f'//*[@id="b061496"]/ul[2]/li[{i}]/div[1]/a').click()
        time.sleep(1)
        
        name =  driver.find_element(By.CLASS_NAME, 'headline').text
        print(name)

        sn.append(name)
        savings[name] = {}
        product_intro = driver.find_element(By.CLASS_NAME, 'info-data3').text     # 제목 밑에 주요 정보 가져오기 
        savings[name]['상품 소개'] = product_intro
        product_info = driver.find_element(By.CSS_SELECTOR, "[id='uiProTabCon1']")
        
        tit_lst = product_info.find_elements(By.CLASS_NAME, 'tit')
        for title in tit_lst:
            st.append(title.text)
            
        
        info_lst =  product_info.find_elements(By.CLASS_NAME, 'infoCont')
        for info in info_lst:
            si.append(info.text)
        
        for t, i in zip(st, si):
            savings[name][t] = i
            
        url = 'https://obank.kbstar.com/quics?page=C016613&cc=b061496:b061496' # 예금 페이지 
        driver.get(url)
        time.sleep(2)

        driver.find_element(By.XPATH, '//*[@id="categoryD00027"]').click() # 적금 페이지
        time.sleep(2)
        
        driver.find_element(By.CSS_SELECTOR, f"[id='pageinput{page}']").click()
        time.sleep(3)
        
        
        
        
            
        

print(len(sn), sn)
print(len(st), st)
print(len(si), si)
print(savings)


########################

# json 파일로 저장
import json
with open(r'C:\Users\campus1N025\Projects\kb_contest\kb_savings.json', 'w') as f : 
	json.dump(savings, f, indent=4,ensure_ascii=False)
```

## 신용카드 크롤링

```python
chrome_options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
driver.implicitly_wait(3)


credit_card = {}
def card_click(tab,row_idx, col_idx):
    col_idx += 1
    driver.find_element(By.XPATH, f'//*[@id="mainForm"]/section[2]/div[{row_idx}]/div[{col_idx}]/div[1]/a').click()
    time.sleep(2)
    
    cardtit = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'cardTit')))
    name = cardtit.find_element(By.CLASS_NAME, 'tit').text
    describe = cardtit.find_element(By.CLASS_NAME, 'txt').text
    credit_card[name] = {}
    credit_card[name]['card_describe'] = describe

    benefit = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME,'benefitList1')))
    li_list = benefit.find_elements(By.TAG_NAME,'li')
    for li in li_list:
        subtitle = li.find_element(By.CLASS_NAME,'tit').text
        text = li.find_element(By.CLASS_NAME,'txt').text
        credit_card[name][subtitle] = {}
        credit_card[name][subtitle] = text 
    driver.back()
    # driver.get(f"https://card.kbcard.com/CRD/DVIEW/HCAMCXPRICAC0047?pageNo=1&cateIdx={tab}")
    time.sleep(2)

credit_page_lst = [1, 3, 2, 5, 6, 10, 11, 7, 8, 4, 9]
check_page_lst = [ 5, 1,  7, 2, 6, 4, 10, 8]
# check card
# https://card.kbcard.com/CRD/DVIEW/HCAMCXPRICAC0056?pageNo=1&cateIdx=5
# credit card
# https://card.kbcard.com/CRD/DVIEW/HCAMCXPRICAC0047?pageNo=1&cateIdx=3

for tab in credit_page_lst:
    driver.get(f"https://card.kbcard.com/CRD/DVIEW/HCAMCXPRICAC0047?pageNo=1&cateIdx={tab}")
# //*[@id="mainForm"]/section[2]/div[2]/div[3]/div[1]/a
# //*[@id="mainForm"]/section[2]/div[2]/div[1]/div[1]/a
# //*[@id="mainForm"]/section[2]/div[8]/div[2]/div[1]/a
    card_boxes= driver.find_elements(By.CLASS_NAME, 'card-box__item')
    card_nums = len(card_boxes)     # 27 
    if card_nums % 3 == 0 :         # 행당 3개
        row_nums = card_nums // 3   # 1  -> 9         
        last_col_nums = 3           # 0
    else:                               # 행당 1개 or 2개 
        row_nums = card_nums // 3 + 1   # 1
        last_col_nums = card_nums % 3 

    for row_idx in range(row_nums):  # 0~8
        row_idx += 2                 # 2~10
        if row_idx  == row_nums + 1 :  # 10 
            for col_idx in range(last_col_nums):
                card_click(tab, row_idx, col_idx)
        else:
            for col_idx in range(3):
                card_click(tab, row_idx, col_idx)
```

```python
chrome_options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
driver.implicitly_wait(3)


check_card = {}
def card_click(tab,row_idx, col_idx):
    col_idx += 1
    driver.find_element(By.XPATH, f'//*[@id="mainForm"]/section[2]/div[{row_idx}]/div[{col_idx}]/div[1]/a').click()
    time.sleep(2)
    
    cardtit = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'cardTit')))
    name = cardtit.find_element(By.CLASS_NAME, 'tit').text
    describe = cardtit.find_element(By.CLASS_NAME, 'txt').text
    check_card[name] = {}
    check_card[name]['card_describe'] = describe

    benefit = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME,'benefitList1')))
    li_list = benefit.find_elements(By.TAG_NAME,'li')
    for li in li_list:
        subtitle = li.find_element(By.CLASS_NAME,'tit').text
        text = li.find_element(By.CLASS_NAME,'txt').text
        check_card[name][subtitle] = {}
        check_card[name][subtitle] = text 
    driver.back()
    # driver.get(f"https://card.kbcard.com/CRD/DVIEW/HCAMCXPRICAC0047?pageNo=1&cateIdx={tab}")
    time.sleep(5)

        
        

# credit_page_lst = [1, 3, 2, 5, 6, 10, 11, 7, 8, 4, 9]
check_page_lst = [5, 1,  7, 2, 6, 4, 10, 8]
# check card
# https://card.kbcard.com/CRD/DVIEW/HCAMCXPRICAC0056?pageNo=1&cateIdx=5
# credit card
# https://card.kbcard.com/CRD/DVIEW/HCAMCXPRICAC0047?pageNo=1&cateIdx=3

for tab in check_page_lst:
    driver.get(f"https://card.kbcard.com/CRD/DVIEW/HCAMCXPRICAC0056?pageNo=1&cateIdx={tab}")
# //*[@id="mainForm"]/section[2]/div[2]/div[3]/div[1]/a
# //*[@id="mainForm"]/section[2]/div[2]/div[1]/div[1]/a
# //*[@id="mainForm"]/section[2]/div[8]/div[2]/div[1]/a
    time.sleep(2)
    card_boxes= driver.find_elements(By.CLASS_NAME, 'card-box__item')
    card_nums = len(card_boxes)     # 9 
    if card_nums % 3 == 0 :         # 행당 3개
        row_nums = card_nums // 3   # 3         
        last_col_nums = 3           # 0
    else:                               # 행당 1개 or 2개 
        row_nums = card_nums // 3 + 1   # 1
        last_col_nums = card_nums % 3 

    for row_idx in range(row_nums):  # 0~3
        if tab == 7:
            row_idx += 2 
            if row_idx == 2:
                for col_idx in range(1,3):
                    card_click(tab, row_idx, col_idx)

            elif row_idx  == row_nums + 1 :  # 10 
                for col_idx in range(last_col_nums):
                    card_click(tab, row_idx, col_idx)

            else:
                for col_idx in range(3):
                    card_click(tab, row_idx, col_idx)
        else:
            row_idx += 2
            if row_idx  == row_nums + 1 :  # 10 
                for col_idx in range(last_col_nums):
                    card_click(tab, row_idx, col_idx)

            else:
                for col_idx in range(3):
                    card_click(tab, row_idx, col_idx)

driver.quit()
```
